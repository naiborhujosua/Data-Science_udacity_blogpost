```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
#show how many nulls value are found in each column 
listings_df.isnull().sum()
```

```python
#drop some columns we find not relevant enough for our analysis in this case are column id, host_name and last_review
columns_to_drop =['id','host_name','last_review']
listings_df.drop(columns_to_drop, axis="columns",inplace=True)
```

```python
listings_df['reviews_per_month']
```

```python
#Because we find that NaN will be difficult to understand in our reviews_per_month column, it is better to change it to 0 for better understandiing.
#we can replace NaN value with this method 
listings_df.fillna({'reviews_per_month':0}, inplace=True)
```

```python
sns.countplot(data=listings_df, x="neighbourhood_group", hue="room_type")
```

```python
#Questions 5: How many flats are more than expensive than 500$/night?Could we remove those outliers from visualization? What techniques can we use?
affordable_df =listings_df[listings_df['price'] <= 500]
sns.distplot(affordable_df['price'])
```

```python
#Questions 6: plots the map of the distribution of flat prices based on neighborhood group?
import urllib
i = urllib.request.urlopen('https://upload.wikimedia.org/wikipedia/commons/e/ec/Neighbourhoods_New_York_City_Map.PNG')
plt.imshow(plt.imread(i), zorder=0, extent=[-74.258, -73.7, 40.49, 40.92])
ax = plt.gca()
affordable_df.plot(
    ax=ax,
    zorder=1,
    kind='scatter',
    x='longitude',
    y='latitude',
    c='price',
    cmap='inferno',
    colorbar=True,
    alpha=0.8,
    figsize=(12,8))
```
